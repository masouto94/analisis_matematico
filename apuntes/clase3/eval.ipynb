```python
from sympy import Matrix, pprint

A_sym = Matrix([
    [1, 3, 2],
    [4, 2, -1],
    [5, 5, 1]
])

for val, mult, vecs in A_sym.eigenvects():
    print(f"λ = {val}, multiplicidad = {mult}")
    for v in vecs:
        pprint(v)

A_sym.eigenvects()
```

```python
from sympy import Matrix

B = Matrix([
    [4, 2, 2],
    [-2,-1,-4],
    [0, 0, 3]
])

for val, mult, vecs in B.eigenvects():
    dim_autoespacio = len(vecs)
    print(f"λ={val}  multiplicidad algebraica={mult}  multiplicidad geométrica={dim_autoespacio}")
    for v in vecs:
        print(f"  {v.T}")
```
