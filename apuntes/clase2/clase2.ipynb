```python
vectors =[ np.array((1,2,1)), np.array((3,0,-1)), np.array((-1,2,-3))] 

def proyeccion_ortogonal(u,v):
    num = np.dot(u,v)
    den = np.dot(u,u)
    return (num/den) * u
def gram_schmitt(vectors):
    ret = []
    for i in range(len(vectors)):
        if not ret:
            ret.append(vectors[i])
            continue
        ret.append(vectors[i] - proyeccion_ortogonal(vectors[i-1], vectors[i]))
    return [v* (1/np.linalg.norm(v)) for v in ret]
print(proyeccion_ortogonal(vectors[0], vectors[1]))
gram_schmitt(vectors)
```
